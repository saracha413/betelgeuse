```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

import os

import sys
print(sys.version)
```

```python
def plot_variations(path):
    files = [path+f for f in os.listdir(path) if (os.path.isfile(os.path.join(path, f))) and (f.find('.DS') == -1) ]
    transits = [np.loadtxt(file, unpack=True, skiprows=2)[1] for file in files]
    #get modal fraction
    #modals = [t/np.sum(t) for t in transits]
    
    wavelengths = np.loadtxt(files[0], unpack=True, skiprows=2)[0]
    
    suffix = '.txt'
    names = [(name.lstrip(path)).rstrip(suffix) for name in files]
    
    fig, axs = plt.subplots(4,3, sharex=True, sharey=True, figsize=(15,10))
    

    
    for i in range(len(transits)):
        if i<3:
            #axs[0,i].plot(wavelengths,modals[i])
            axs[0,i].plot(wavelengths,transits[i])
            axs[0,i].set_title(names[i], y=0.8)
        elif i<6:
            #axs[1,i-3].plot(wavelengths,modals[i])
            axs[1,i-3].plot(wavelengths,transits[i])
            axs[1,i-3].set_title(names[i], y=0.8)
        elif i<9:
            #axs[2,i-6].plot(wavelengths,modals[i])
            axs[2,i-6].plot(wavelengths,transits[i])
            axs[2,i-6].set_title(names[i], y=0.8)
        else:
            #axs[3,i-9].plot(wavelengths,modals[i])
            axs[3,i-9].plot(wavelengths,transits[i])
            axs[3,i-9].set_title(names[i], y=0.8)
            
    for ax in axs.flat:
        ax.set_xlabel('wavelengths(um)')
        #ax.set_ylabel('modal fraction')
        ax.set_ylabel('raw spectrum')
    
    #plt.legend()
    #plt.xlabel('wavelengths (um)')
    #plt.ylabel('Modal fraction')
    
```

```python
jup_cloud_path = '/Users/saravannah/betelgeuse/single_param_jup_variants/jup_clouds/'
plot_variations(jup_cloud_path)

#varying cloud cover
plt.suptitle('Varying cloud cover',fontsize=20)
plt.savefig('varying_cloud_cover.png')
```
