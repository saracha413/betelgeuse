```python
import matplotlib.pyplot as plt
import numpy as np

from spectools_ir.spectools_ir.utils import extract_hitran_data

%matplotlib inline
```

```python
xlims=[1e-4, 15]#[4.64,4.92]
threshold_S = 3e-19

#print(max(H2O_raw['sw']))
```

```python
#does this also get me the multiplet width??? Can use to get bins
H2O_raw=extract_hitran_data('H2O',xlims[0],xlims[1],vup=1)  #Outputs an astropy table

H2O={'wave':[],'Qpp':[]}
for i in range(len(H2O_raw['wave'])):
    if H2O_raw['sw'][i] > threshold_S:
        H2O['wave'].append(H2O_raw['wave'][i])
        H2O['Qpp'].append(H2O_raw['Qpp'][i])
```

```python
#does this also get me the multiplet width??? Can use to get bins
CO_raw=extract_hitran_data('CO',xlims[0],xlims[1],vup=1)  #Outputs an astropy table

CO={'wave':[],'Qpp':[]}
for i in range(len(CO_raw['wave'])):
    if CO_raw['sw'][i] > threshold_S:
        CO['wave'].append(CO_raw['wave'][i])
        CO['Qpp'].append(CO_raw['Qpp'][i])
```

```python
print(CO_raw.colnames) #sw is the spectral line intensity
```

```python
earth_infile = '/Users/saravannah/AstroSpec/Exo_Transmit_spectra/earth.txt'
earth_wave, earth_flux = np.loadtxt(earth_infile, unpack=True, skiprows=2)
earth_wave = earth_wave * 1e6
```

```python
for i,mywave in enumerate(CO['wave']):
    plt.axvline(mywave,color='C1')
    plt.text(CO['wave'][i],0.02,CO['Qpp'][i].strip())
        
plt.plot(earth_wave, earth_flux)
plt.xlim(xlims)
plt.gcf().set_size_inches(10, 5)
```
