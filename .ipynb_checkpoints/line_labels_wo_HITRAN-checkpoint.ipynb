```python
def find_linecenter(wavelength, intensity):
    
    line_IDs = []
    
    max_intensity = 0
    peak_wavelength = 0
    for i in range(len(wavelength)):
        if intensity[i] > 0:
            if intensity[i] > max_intensity:
                peak_wavelength = wavelength[i]
                max_intensity = intensity[i]
                
        #intensity is 0 but peak intensity isn't and last intensity wasn't 0 
        elif max_intensity != 0: 
            #save peak of intensity and wavelength of that peak as line ID
            line_IDs.append(peak_wavelength)
            max_intensity, peak_wavelength = 0,0
            
    return line_IDs
    
            
    

            
            
```
