```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft, fftfreq, rfft, rfftfreq
from scipy.interpolate import interp1d
```

```python
earth = np.loadtxt('/Users/saravannah/Exo_Transmit_BACKUP/Spectra/earth.dat', unpack=True, skiprows=2)
jup = np.loadtxt('/Users/saravannah/Exo_Transmit_BACKUP/Spectra/changingCompT=300/1Xsolar_cond_Jupiter.dat', unpack=True, skiprows=2)
#hot Jupiter with pronounced K and Na lines
wasp = np.loadtxt('/Users/saravannah/Exo_Transmit_BACKUP/Spectra/hotJupiters_NEW/WASP-39b.dat', unpack=True, skiprows=2)
```

```python
#interpolation
earth_interp = interp1d(earth[0],earth[1],kind='linear')
jup_interp = interp1d(jup[0],jup[1],kind='linear')
wasp_interp = interp1d(wasp[0],wasp[1],kind='linear')
```

```python
strong_CO = [4.65, 5.02]
strong_H2O = [0.94, 1.45, 1.77, 2.52, 2.85, 5.91, 6.68, 7.60, 9.19, 13.26, 17.43, 19.91]
strong_CO2 = [4.27, 14.35]
strong_Na = [0.59]
strong_CH4 = [3.43]
strong_K = [0.40, 0.77]
```

```python
plt.plot(earth[0]*1e6, earth[1], label='Earth')
plt.plot(jup[0]*1e6, jup[1], label='Jupiter')
plt.plot(wasp[0]*1e6, wasp[1], label='WASP-39b')

x = np.arange(min(earth[0]), max(earth[0]), 1e-8)

plt.plot(x*1e6, earth_interp(x), label='Earth interp')
plt.plot(x*1e6, jup_interp(x), label='Jupiter interp')
plt.plot(x*1e6, wasp_interp(x), label='WASP-39b interp')


#size of left shift to sent absorption labels over the line
centering = 0.04

#how high up on y-axis to put text to label absorption line
height = 2.8

#plot absorption line labels
for i in range(len(strong_CO)):
    plt.axvline(x=strong_CO[i], ymin = 0.7, ymax=0.9, color='black')
    plt.text(strong_CO[i] - centering*strong_CO[i], height, 'CO', fontsize='small')
for i in range(len(strong_H2O)):
    plt.axvline(x=strong_H2O[i], ymin = 0.7, ymax=0.9, color='blue', linestyle='dashed')
    plt.text(strong_H2O[i] - centering*strong_H2O[i],  height, 'H2O', fontsize='small')
for i in range(len(strong_CO2)):
    plt.axvline(x=strong_CO2[i], ymin = 0.7, ymax=0.9, color='green')
    plt.text(strong_CO2[i] - centering*strong_CO2[i],  height, 'CO2', fontsize='small')
for i in range(len(strong_Na)):
    plt.axvline(x=strong_Na[i], ymin = 0.7, ymax=0.9, color='red')
    plt.text(strong_Na[i] - centering*strong_Na[i]*0.7,  height, 'Na', fontsize='small')
for i in range(len(strong_CH4)):
    plt.axvline(x=strong_CH4[i], ymin = 0.7, ymax=0.9, color='orange')
    plt.text(strong_CH4[i] - centering*strong_CH4[i],  height, 'CH4', fontsize='small')
for i in range(len(strong_K)):
    plt.axvline(x=strong_K[i], ymin = 0.7, ymax=0.9, color='teal')
    plt.text(strong_K[i] - centering*strong_K[i]*0.4,  height, 'K', fontsize='small')

plt.xlabel('wavelength (um)')
plt.ylabel('percent absoption (%)')
plt.xscale('log')
plt.legend()
plt.title('Raw spectra')

plt.gcf().set_size_inches(12, 7)
```

```python
#following example in https://docs.scipy.org/doc/scipy/tutorial/fft.html#and-n-d-discrete-fourier-transforms

#get wavenumbers
# Number of sample points
N = len(earth[0]) #note all wavelength arrays have the same values
# sample spacing
T = (max(earth[0])-min(earth[0]))/N

xf = rfftfreq(N, T)



plt.scatter(xf, rfft(earth[1]), label='Earth')
plt.scatter(xf, rfft(jup[1]), label='Jupiter')
plt.scatter(xf, rfft(wasp[1]), label='WASP-39b')


#plt.scatter(wavenumbers, rfft(earth[1]), label='Earth')
#plt.scatter(wavenumbers, rfft(jup[1]), label='Jupiter')
#plt.scatter(wavenumbers, rfft(wasp[1]), label='WASP-39b')

#size of left shift to sent absorption labels over the line
centering = 0.04

#how high up on y-axis to put text to label absorption line
height = 2.8

plt.xlabel('wavelength (um)')
plt.ylabel('fft(percent absoption (%))')
plt.xscale('log')
plt.yscale('log')
plt.legend()

plt.title('Fourier transforms')

plt.gcf().set_size_inches(10, 5)

```

```python
def get_modal(absorptions):
    power = rfft(absorptions) #no special name for fft or absorption %
    
    #normalize
    modal = power*np.conj(power)/sum(power*np.conj(power))
    #force Python to drop the "+0j" part of the "complex" number
    modal = np.real(modal) 
    
    return modal
```

```python
fig, axs = plt.subplots(3,1, sharex=True, sharey=True, figsize=(12,10))

axs[0].plot(xf, get_modal(earth_interp(x)), label='Earth')
axs[1].plot(xf, get_modal(jup_interp(x)), label='Jupiter')
axs[2].plot(xf, get_modal(wasp_interp(x)), label='WASP')

for ax in axs:
    ax.set_yscale('log')
    ax.legend()
    ax.set_xlabel('wavenumber')
    ax.set_ylabel('Modal fraction')
    #plt.ylim([0,50])
```

```python
#p and q are modal fractions for the two planets
def djs_density(p, q):
    r = 1/2 * (p+q)
    #p,r,q = np.ma.array(p), np.ma.array(r), np.ma.array(q)
    djs_dens = 1/2 * p*np.log(p/r) + 1/2 * q*np.log(q/r)
    
    return djs_dens
```

```python
wasp_earth_dens = djs_density(get_modal(earth_interp(x)), get_modal(wasp_interp(x)))
wasp_jup_dens = djs_density(get_modal(jup_interp(x)), get_modal(wasp_interp(x)))


fig, axs = plt.subplots(2,1, sharex=True, sharey=True, figsize=(12,10))

axs[0].plot(xf, wasp_earth_dens, label='WASP vs Earth')
axs[1].plot(xf, wasp_jup_dens, label='WASP vs. Jupiter')

                              
for ax in axs:
    ax.set_yscale('log')
    ax.legend()
    ax.set_xlabel('wavenumber')
    ax.set_ylabel('Djs density')
```

```python
print(sum(wasp_earth_dens), sum(wasp_jup_dens))
```

```python
#information bias
earth_bias = np.log2(np.exp(1))/(2*sum(fft(earth[1])))*(max(xf)-min(xf)-1)
jup_bias = np.log2(np.exp(1))/(2*sum(fft(jup[1])))*(max(xf)-min(xf)-1)


print(earth_bias, jup_bias)
```
