```python
#### 
temps = [300,400,500,600,700] #temperatures in K
dir_names = ["changingCompT="+str(i) for i in temps] #directories named by temperature
metallicities = ['0p1', '1', '5', '30', '50', '100', '1000']

### change to where your exo transmit directory is stored
d = [['/Users/saravannah/Exo_Transmit_BACKUP/Spectra/changingCompT='+str(T)+'/'+X+'Xsolar_cond_Jupiter.dat' for T in temps] for X in metallicities]
fileExtensions = pd.DataFrame(data=d, index=metallicities, columns=temps)
Jupiter = np.loadtxt(fileExtensions[300][1], unpack=True, skiprows=2)
Earth = np.loadtxt('/Users/saravannah/Exo_Transmit_BACKUP/Spectra/earth.dat', unpack=True, skiprows=2)

Earth_without_water = np.loadtxt('/Users/saravannah/Exo_Transmit_BACKUP/Spectra/earth_without_water.dat', unpack=True, skiprows=2)
```

```python
#what do spectra look like? 

for T in temps:
    for X in metallicities:
        spectrum=np.loadtxt(fileExtensions[T][X], unpack=True, skiprows=2)
        plt.plot(spectrum[0], spectrum[1])
#wow! absolutely metal. Would probably be easier to analyze if they weren't all on top of each other, though
```

```python
fig, axs = plt.subplots(7, 1, sharex=True, sharey=True)

for T in temps:
    for X in metallicities:
        spectrum=np.loadtxt(fileExtensions[T][X], unpack=True, skiprows=2)
        axs[metallicities.index(X)].plot(spectrum[0]*1e6, spectrum[1], label='T='+str(T)+'K')
        
for i in range(len(metallicities)):
    axs[i].legend()
    axs[i].set_title(metallicities[i]+' times solar metallicity')
    
    axs[i].set_ylabel('Percent absorption (%)')
    axs[i].set_xscale('log')
axs[6].set_xlabel('wavelength (um)')
plt.gcf().set_size_inches(12, 15)
```

```python
def get_Djs(fileExtension, comparison_planet_flux):
    wavelength, flux = np.loadtxt(fileExtension, unpack=True, skiprows=2)
    
    return djs(flux, comparison_planet_flux)
    
djs_Jupiter = fileExtensions.applymap(lambda x: get_Djs(x,Jupiter[1]))
djs_Earth = fileExtensions.applymap(lambda x: get_Djs(x,Earth[1]))
djs_Earth_without_water = fileExtensions.applymap(lambda x: get_Djs(x,Earth_without_water[1]))
```

```python
#plot with broken y-axis to show difference between Earth and Jupiter Djs
#instructions from Matplotlib documentation:
#https://matplotlib.org/stable/gallery/subplots_axes_and_figures/broken_axis.html

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
fig.subplots_adjust(hspace=0.05)  # adjust space between axes

# plot the same data on both axes

ax1.scatter(djs_Earth.index, djs_Earth[300], label='T=300, Earth', color=cmap_Earth.to_rgba(300), marker='*')
ax1.scatter(djs_Earth.index, djs_Earth[400], label='T=400, Earth', color=cmap_Earth.to_rgba(400), marker='*')
ax1.scatter(djs_Earth.index, djs_Earth[500], label='T=500, Earth', color=cmap_Earth.to_rgba(500), marker='*')
ax1.scatter(djs_Earth.index, djs_Earth[600], label='T=600, Earth', color=cmap_Earth.to_rgba(600), marker='*')
ax1.scatter(djs_Earth.index, djs_Earth[700], label='T=700, Earth', color=cmap_Earth.to_rgba(700), marker='*')

ax2.scatter(djs_Jupiter.index, djs_Jupiter[300], label='T=300, Jupiter', color=cmap_Jupiter.to_rgba(300))
ax2.scatter(djs_Jupiter.index, djs_Jupiter[400], label='T=400, Jupiter', color=cmap_Jupiter.to_rgba(400))
ax2.scatter(djs_Jupiter.index, djs_Jupiter[500], label='T=500, Jupiter', color=cmap_Jupiter.to_rgba(500))
ax2.scatter(djs_Jupiter.index, djs_Jupiter[600], label='T=600, Jupiter', color=cmap_Jupiter.to_rgba(600))
ax2.scatter(djs_Jupiter.index, djs_Jupiter[700], label='T=700, Jupiter', color=cmap_Jupiter.to_rgba(700))

# zoom-in / limit the view to different portions of the data
ax1.set_ylim(8e-3, 9e-3)  # outliers only
ax2.set_ylim(0, 1e-3)  # most of the data

# hide the spines between ax and ax2
ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax1.xaxis.tick_top()
ax1.tick_params(labeltop=False)  # don't put tick labels at the top
ax2.xaxis.tick_bottom()

# This looks pretty good, and was fairly painless, but you can get that
# cut-out diagonal lines look with just a bit more work. The important
# thing to know here is that in axes coordinates, which are always
# between 0-1, spine endpoints are at these locations (0,0), (0,1),
# (1,0), and (1,1).  Thus, we just need to put the diagonals in the
# appropriate corners of each of our axes, and so long as we use the
# right transform and disable clipping.

d = .015  # how big to make the diagonal lines in axes coordinates
# arguments to pass to plot, just so we don't keep repeating them
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal

ax1.set_yscale('log')
ax2.set_yscale('log')

# What's cool about this is that now if we vary the distance between
# ax and ax2 via f.subplots_adjust(hspace=...) or plt.subplot_tool(),
# the diagonal lines will move accordingly, and stay right at the tips
# of the spines they are 'breaking'

ax2.set_xlabel('Multiple of solar metallicity')
ax1.set_ylabel('Djs')

ax1.axvline(x=1, color='gray', linestyle='dashed')
ax2.axvline(x=1, color='gray', linestyle='dashed')

plt.gcf().set_size_inches(10, 7)
ax1.legend()
ax2.legend()
```
